```python
#OASIS EDA
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
```

```python
dir_path = '../Dataset/'
file_name = 'combined_data.csv'
csv_path = os.path.join(dir_path,file_name)
```

```python
df_org = pd.read_csv(csv_path)
```

```python
df_org.head(10)
```

```python
df_base = df_org[['source','subject','MR_ID','group_maxinc','sex','age']]
df_base.head(5)
```

```python
oasis_nii = os.listdir('../Dataset/OASIS_BET/')
adni_nii = os.listdir('../Dataset/ADNI_BET/')
```

```python
df_oasis = pd.DataFrame(oasis_nii,columns=['filename'])
df_adni = pd.DataFrame(adni_nii,columns=['filename'])
```

```python
df_oasis['path'] = "Dataset/OASIS_BET/"+df_oasis['filename']
df_adni['path'] = "Dataset/ADNI_BET/"+df_adni['filename']
```

```python
df_oasis['MR_ID'] = df_oasis['path'].str[22:].replace('_ses-','_MR_',regex=True).replace('_sess-','_MR_',regex=True).str[:17]
df_adni['MR_ID'] = df_adni['path'].str[17:].replace('.nii','',regex=True)
```

```python
df_dataset = pd.concat([df_oasis,df_adni])
```

```python
df_dataset_final = pd.merge(left=df_base,right=df_dataset,how='left',on='MR_ID')
df_dataset_final.head(5)
```

```python
df_dataset_final.to_csv('../Dataset/Dataset_with_path.csv',index=False)
```

```python
df_min = df_dataset_final.groupby(['source','subject','MR_ID','group_maxinc','sex','age']).min().reset_index()
```

```python
df_min.to_csv('../Dataset/Dataset_min.csv',index=False)
```
