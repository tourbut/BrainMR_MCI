```python
#OASIS EDA
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
```

```python
dir_path = '../Dataset/'
file_name = 'combined_data.csv'
csv_path = os.path.join(dir_path,file_name)
```

```python
df_org = pd.read_csv(csv_path)
```

```python
df_org.head(10)
```

```python
df_base = df_org[['source','subject','MR_ID','group_maxinc','sex','age']]
df_base.head(5)
```

```python
df_base['source'].value_counts()
```

```python
df_base
```

```python
oasis_nii = glob.glob('../Dataset/OASIS_BET/*.nii')
adni_nii = glob.glob('../Dataset/ADNI_BET/*.nii')
```

```python
df_oasis = pd.DataFrame(oasis_nii,columns=['path'])
df_adni = pd.DataFrame(adni_nii,columns=['path'])
```

```python
df_oasis
```

```python
df_oasis['MR_ID'] = df_oasis['path'].str.replace('../Dataset/OASIS_BET/sub-','',regex=True).replace('_ses-','_MR_',regex=True).replace('_sess-','_MR_',regex=True).str[:17]
df_oasis['filename'] = df_oasis['path'].str.replace('../Dataset/OASIS_BET/','',regex=True)
```

```python
df_adni['MR_ID'] = df_adni['path'].str.replace('../Dataset/ADNI_BET/','',regex=True).replace('.nii','',regex=True)
df_adni['filename'] = df_adni['path'].str.replace('../Dataset/ADNI_BET/','',regex=True)
```

```python
df_dataset = pd.concat([df_oasis,df_adni])
df_dataset
```

```python
df_dataset_final = pd.merge(left=df_base,right=df_dataset,how='left',on='MR_ID')
df_dataset_final.head(5)
```

```python
df_dataset_final.to_csv('Dataset_with_path.csv',index=False)
```
